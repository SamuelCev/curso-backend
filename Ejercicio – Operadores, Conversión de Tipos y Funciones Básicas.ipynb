```python
#Calculadora

base = int(input("Ingresa el sistema numerico con el que deseas trabajar\n1 - Decimal\n2 - Binario\n3 - Octal\n4 - Hexadecimal\nSelección: "))
print("\n")
lista = [input(f"Ingresa el valor #{i + 1} de la lista: ") for i in range(5)]
Set = {input(f"Ingresa el valor #{i + 1} del set: ") for i in range(5)}
dic = {i : input(f"Ingresa el valor #{i + 1} del diccionario: ") for i in range(5)}
oper = int(input("\nIngresa la operacion que quieres realizar\n1 - Suma\n2 - Resta\n3 - Multiplicación\n4 - División\n5 - Potenciación\nSelección: "))

if len(lista) == 5 and len(Set) == 5 and len(dic) == 5:
    try:
        if base == 1:
            lista = [float(x) for x in lista]
            Set = {float(x) for x in Set}
            dic = {x: float(dic[x]) for x in dic}
        elif base == 2:
            lista = [int(x,2) for x in lista]
            Set = {int(x,2) for x in Set}
            dic = {x: int(dic[x],2) for x in dic}
        elif base == 3:
            lista = [int(x,8) for x in lista]
            Set = {int(x,8) for x in Set}
            dic = {x: int(dic[x],8) for x in dic}
        elif base == 4:
            lista = [int(x,16) for x in lista]
            Set = {int(x,16) for x in Set}
            dic = {x: int(dic[x],16) for x in dic}
        else:
            print("Error en el sistema númerico seleccionado.")

        resLista = lista[0]
        resSet = None
        resDic= dic[0]
        
        print("\nLos valores de la lista son: ", lista)
        print("Los valores de la lista son: ", Set)
        print("Los valores de la lista son: ", dic)
        print("\n")
        
        if oper == 1:
            print('La operación seleccionada fue "Suma".')
            print("Suma de la lista: ",sum(lista))
            print("Suma del set: ", sum(Set))
            print("Suma del dicccionario: ", sum(dic.values()))
        elif oper == 2:
            print('La operación seleccionada fue "Resta".')
            for i in range(1,5):
                resLista -= lista[i]
            for num in Set:
                if resSet is None:
                    resSet = num
                else:
                    resSet -= num
            for i in range(1,5):
                resDic -= dic[i]
            print("Resta de la lista: ",resLista)
            print("Resta del set: ", resSet)
            print("Resta del dicccionario: ", resDic)
        elif oper == 3:
            print('La operación seleccionada fue "Multiplicación".')
            for i in range(1,5):
                resLista *= lista[i]
            for num in Set:
                if resSet is None:
                    resSet = num
                else:
                    resSet *= num
            for i in range(1,5):
                resDic *= dic[i]
            print("Multiplicación de la lista: ",resLista)
            print("Multiplicación del set: ", resSet)
            print("Multiplicación del dicccionario: ", resDic)
        elif oper == 4:
            print('La operación seleccionada fue "División".')
            for i in range(1,5):
                resLista /= lista[i]
            for num in Set:
                if resSet is None:
                    resSet = num
                else:
                    resSet /= num
            for i in range(1,5):
                resDic /= dic[i]
            print("División de la lista: ",resLista)
            print("División del set: ", resSet)
            print("División del dicccionario: ", resDic)
        elif oper == 5:
            print('La operación seleccionada fue "Potencia".')
            for i in range(1,5):
                resLista **= lista[i]
            for num in Set:
                if resSet is None:
                    resSet = num
                else:
                    resSet **= num
            for i in range(1,5):
                resDic **= dic[i]
            print("Potencia de la lista: ",resLista)
            print("Potencia del set: ", resSet)
            print("Potencia del dicccionario: ", resDic)
        else:
            print("Error en la operación seleccionada.")

    except:
        print("Se introdujo algun caracter que no es un numero.")
else:
    print("Entrada invalida")
```
